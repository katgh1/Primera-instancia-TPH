```python
#importo lo necesario para limpieza
import nltk
from nltk.corpus import stopwords
import re
nltk.download('stopwords')
stopwords_es = set(stopwords.words('spanish'))
```

```python
# Funcion para limpiar texto, será util para cada álbum

def limpiar_texto(texto):

  texto_minuscula = texto.lower() # convierte todo el texto en minúsculas

  texto_sin_puntuacion = re.sub(r'[^\w\s]', '', texto_minuscula) # elimina todos los signos de puntuacion

  lista_palabras = texto_sin_puntuacion.split() # divide el texto en palabras individuales

  palabras_importantes = [] # crea lista de palabras importantes

  for palabra in lista_palabras:
    if palabra not in stopwords_es:
      palabras_importantes.append(palabra)

  return palabras_importantes
```

```python
def limpiar_texto(texto):
  #Defino una lista de palabras para agregar a las stopwords
    palabras_a_eliminar = ['dijo', 'así', 'aunque', 'sino', 'luego', 'pues', 'mientras',
        'después', 'antes', 'porque', 'cuando', 'cómo', 'donde', 'cap',
        'capítulo', 'verso', 'canto', 'á', 'ó', 'si' , 'tan' , 'fué', 'ves' , 'hacia' ]

    black_list = stopwords_es.union(palabras_a_eliminar) #uno stopwords y palabras extra
    texto_minuscula = texto.lower() #texto a minuscula
    texto_sin_puntuacion = re.sub(r'[^\w\s]', '', texto_minuscula) #elimino signos de puntuacion
    lista_palabras = texto_sin_puntuacion.split() #Separa al texto en palabras

    palabras_importantes = [palabra for palabra in lista_palabras if palabra not in black_list]

    return palabras_importantes
```

```python
from collections import Counter
#from nltk.corpus import PlaintextCorpusReader
from nltk.probability import ConditionalFreqDist
```

```python
# 1. Unimos todas las listas
todas_las_palabras = letras_YDLCAL_limpias + CM_limpias + PB_limpias + SNM_limpias


conteo = Counter(todas_las_palabras)

print(" Palabras mas repetidas: ", conteo.most_common(12))
```

```python
# Lista de álbumes con sus listas de palabras ya limpias
albumes = [
    ('Yendo de la cama al living', letras_YDLCAL_limpias),
    ('Clics modernos', CM_limpias),
    ('Piano Bar', PB_limpias),
    ('Say No More', SNM_limpias)
]

# Palabras que querés comparar entre los álbumes
palabras_clave = ['amor','mundo', 'desaparecer']

# Crear la lista de tuplas (palabra, álbum) solo si la palabra está en las palabras clave
tuplas = [
    (palabra, album)
    for album, lista in albumes
    for palabra in lista
    if palabra in palabras_clave
]

# Crear una frecuencia condicional de esas tuplas
cfd = ConditionalFreqDist(tuplas)

# Graficar el resultado
plt.figure(figsize=(10,5))
cfd.plot(title="Frecuencia de palabras clave por álbum")
```

##Por la estructura que suelen seguir las canciones el hecho de que una palabra se repita una única vez tambien da pistas sobre el contexto, ya que son palabras que aparecen sólo en un álbum ¿Serán indicadores de algo?


```python
# Filtrar solo las palabras que aparecen una vez
palabras_unicas = [palabra for palabra, frecuencia in conteo.items() if frecuencia == 1]

print("Palabras que aparecen solo una vez  :")
print(palabras_unicas)
```
